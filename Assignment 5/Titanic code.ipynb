```python
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x = df[["Pclass", "Sex", "Age"]]
```

```python
y = df.Survived
```

```python
x.isnull().sum()
```

```python
x.Age.fillna(x.Age.mean(), inplace = True)
```

```python
x.isnull().sum()
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder  
le = LabelEncoder() 
gen = le.fit_transform(x.Sex) 
```

```python
x["Gender"] = gen
x.drop("Sex", axis = 1, inplace = True)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Survived', grid=False)
plt.title('Survival by Age')
plt.suptitle('')  # Suppress the default title to keep the plot clean
plt.xlabel('Survived')
plt.ylabel('Age')
plt.xticks([1, 2], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Survived'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
survived_by_gender = df.groupby('Sex')['Survived'].value_counts(normalize=True).unstack()
survived_by_gender.plot(kind='bar', stacked=True, color=['blue', 'orange'])
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
